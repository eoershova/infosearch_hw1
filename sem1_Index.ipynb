### CountVectorizer

````python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
 
# инициализируем
vectorizer = CountVectorizer()

# составляем корпус документов
corpus = [
  'слово1 слово2 слово3',
  'слово2 слово3',
  'слово1 слово2 слово1',
  'слово4'
]

# считаем
X = vectorizer.fit_transform(corpus)
 
# получится следующая структура:
#        | слово1 | слово2 | слово3 | слово4
# текст1 |   1    |    1   |   1    |   0
# текст2 |   0    |    1   |   1    |   0
# текст3 |   2    |    1   |   0    |   0
# текст4 |   0    |    0   |   0    |   1
 
# чтобы получить сгенерированный словарь, из приведенной структуры CountVectorizer
# порядок совпадает с матрицей
vectorizer.get_feature_names()  # ['слово1', 'слово2', 'слово3', 'слово4']
 
# чтобы узнать индекс токена в словаре
vectorizer.vocabulary_.get('слово3') # вернет 2
 
# показать матрицу
X.toarray()
 
# теперь можно быстро подсчитать вектор для нового документа
vectorizer.transform(["слово1 слово4 слово4"])  # результат [[1 0 0 2]]
 
# чтобы узнать количественное вхождение каждого слова:
matrix_freq = np.asarray(X.sum(axis=0)).ravel()
final_matrix = np.array([np.array(vectorizer.get_feature_names()), matrix_freq])
````

##  Индекс 

Сам по себе индекс - это просто формат хранения данных, он не может осуществлять поиск. Для этого необходимо добавить к нему определенную метрику. Это может быть что-то простое типа булева поиска, а может быть что-то более специфическое или кастомное под задачу.

Давайте посмотрим, что полезного можно вытащить из самого индекса.    
По сути, индекс - это информация о частоте встречаемости слова в каждом документе.   
Из этого можно понять, например:
1. какое слово является самым часто употребимым / редким
2. какие слова встречаются всегда вместе - так можно парсить твиттер, fb, форумы и отлавливать новые устойчивые выражения в речи
3. как эти документы кластеризуются по N тематикам согласно словам, которые в них упоминаются

````python
import re
import json
import pickle
````

````python
friends_name_normalization = [
            {'name forms regex': re.compile(r'Мо(?:н|ник)\S*'), 'normalized name': 'Моника'},
            {'name forms regex': re.compile(r'Р(?:э|е)чел\S*'), 'normalized name': 'Рейчэл'},
            {'name forms regex': re.compile(r'Ч(?:э|е)ндлер\S*'), 'normalized name': 'Чендлер'},
            {'name forms regex': re.compile(r'Фиб(?:и|с)'), 'normalized name': 'Фиби'},
            {'name forms regex': re.compile(r'Рос\S{1,3}'), 'normalized name': 'Росс'},
            {'name forms regex': re.compile(r'Джо(?:уи|и)\S*'), 'normalized name': 'Джоуи'}
        ]
````

````python
from pymystem3 import Mystem
m = Mystem()
````

````python
def preproc(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    text = re.sub(r'9999\n00:00:0,500 --> 00:00:2,00\nwww.tvsubtitles.net', '', text)
    text = re.sub(r'movie info: XVID  464x352 25.0fps 99.7 MB\n/SubEdit b.4056 (http://subedit.com.pl)/', ' ', text)
    text = re.sub(r'\ufeff', ' ', text)
    text = re.sub(r'\.\.\.\n\.\.\.', ' ', text)
    text = re.sub(r'\.\.\.', '.', text)
    text = re.sub(r'[?!.]', ' ', text)
    text = re.sub(r'[№%,":;=()]', ' ', text)
    text = re.sub(r'\s-\s', ' ', text)
    text = re.sub(r'\n', ' ', text)
    sentences = text.split('.')
    friends = ['Моника', 'Рейчел', 'Чендлер', 'Фиби', 'Росс', 'Джоуи']
    for friend in friends_name_normalization:
            text = re.sub(friend['name forms regex'], friend['normalized name'], text)
    proc_words = []
    words = text.split()
    names = []
    for word in words:
        if word not in friends:
            proc_words.append(word.lower())
        else:
            names.append(word)
    words = " ".join(proc_words)
    lemmas = m.lemmatize(words)
    lemmas.extend(names)
    words = " ".join(lemmas)
    return words
````

````python
def find_files(data_folder):
    file_paths = []
    for root, dirs, files in os.walk(data_folder):
        for name in files:
            if name != '.DS_Store':
                file_paths.append(os.path.join(root, name))
    return file_paths
````

````python
%%time
corpus = []
episodes = []
files = find_files('friends-data')
for file in files:
    words = preproc(file)
    episodes.append(file.split('/')[-1])
    corpus.append(words)
    
````

````python
vectorizer = CountVectorizer()
````

````python
X = vectorizer.fit_transform(corpus)
````

````python
with open('term_doc_matrix', 'wb') as file:
    pickle.dump(X, file)
````

````python
matrix = X.toarray()
````

````python
matrix.T
````

````python
terms = vectorizer.get_feature_names()
````

````python
from collections import defaultdict
d = defaultdict(list)
````

````python
%%time
for term, matrix_row in zip(terms, matrix.T):
    for count, episode in zip(matrix_row, episodes):
        if count > 0:
            d[term].append(episode)
````

````python
import json
````

````python
with open('friends_index.json', 'w', encoding='utf-8') as file:
    file.write('[\n')
    for key, value in d.items():
        entry = {key: value}
        json.dump(entry, file, indent=4, ensure_ascii=False)
        file.write(',\n')
    file.write('\n]')
````
